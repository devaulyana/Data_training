```python
import pandas as pd 
import wittgenstein as lw
hongos_df = pd.read_csv(r'C:\Users\Ульяна\Desktop\jupiter\datasets\mushrooms.csv')
hongos_df.head(10)
```

```python
hongos_df.info() #Las columnas contienen datos categóricos y se pueden convertir para fines de optimización
```

```python
for col in hongos_df.columns:
    hongos_df[col] = hongos_df[col].astype('category')
    print(f'columna {col}, únicos - {len(hongos_df[col].unique())}')
```

```python
# Ahora el marco de datos ocupa casi 10 veces menos espacio
hongos_df.info()
```

```python
# El paquete wittgenstein implementa las clases IREP y RIPPER que representan los algoritmos del mismo nombre
# 1) prune_size - para un conjunto de entrenamiento que participa en la etapa de reducción (ajuste) de una regla obtenida con otras muestras de conjunto de entrenamiento
# 2) max_rules - número máximo de reglas
# 3) random_state - un inicializador de valores aleatorios
# 4) max_rule_conds - número máximo de condiciones en una regla

# Cree un objeto clasificador y divida los datos en una muestra de entrenamiento y prueba
```

```python
# clf = lw.IREP()
clf = lw.RIPPER()
from sklearn.model_selection import train_test_split
hongos_df_tr, hongos_df_ts = train_test_split(hongos_df, test_size=0.2)
```

```python
clf.fit(hongos_df_tr, class_feat='class', pos_class='p')
#clf.fit(hongos_df_tr.drop('class', axis=1), hongos_df_tr['class'].map({'p': 1, 'e': 0}).astype(int))
```

```python
y_pr = clf.predict(hongos_df_ts)
clf.predict(hongos_df_ts, give_reasons=True)[1][:5]
```

```python
from sklearn.metrics import classification_report
print(clf.score(hongos_df_ts, y_pr, score_function=classification_report))
```

```python
print(classification_report(hongos_df_ts['class'].map({'p':True, 'e':False}).astype(bool), y_pr))
```

```python
clf.out_model() #En nuestro caso, el clasificador con 100% de precisión forma 8 reglas para determinar hongos venenosos, el resto son comestibles. Ahora puedes ir al bosque para recoger setas sin miedo!
```
