```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
houses = pd.read_csv(r'C:\Users\Ульяна\Desktop\jupiter\datasets\1.4_houses.csv')
```

```python
# Visualización de datos
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()
ax.scatter(houses["dim_1"], houses["price"], s=10)

fig = plt.figure(figsize=(5, 5))
markers = {"basic": "o", "medium": "v", "luxury": "*"}
for d1, d2, l in zip(houses.dim_1, houses.dim_2, houses.level):
    plt.scatter(d1, d2, s=20, marker=markers[l])
plt.show()
```

```python
# Agrupación de datos
from sklearn.cluster import KMeans

X = houses[["dim_1", "dim_2"]]
clust = KMeans(n_clusters=3).fit(X)
[c1, c2, c3] = clust.cluster_centers_
```

```python
# Visualización de los resultados de la agrupación
fig = plt.figure(figsize=(10, 10))  # Не добавляйте параметр facecolor для черного фона
markers = {"basic": "o", "medium": "v", "luxury": "*"}
for d1, d2, l in zip(houses.dim_1, houses.dim_2, houses.level):
    plt.scatter(d1, d2, s=250, marker=markers[l])

plt.scatter(c1[0], c1[1], s=250, marker="x", c="white")
plt.scatter(c1[0], c1[1], s=250 * 1e2, c="white", alpha=0.1)

plt.scatter(c2[0], c2[1], s=250, marker="x", c="white")
plt.scatter(c2[0], c2[1], s=250 * 1e2, c="white", alpha=0.1)

plt.scatter(c3[0], c3[1], s=250, marker="x", c="white")
plt.scatter(c3[0], c3[1], s=250 * 3.5e2, c="white", alpha=0.1)

plt.show()
```

```python
# **Resultados
# - A partir de la visualización de los datos, se identificaron 3 grupos de viviendas.
# - Mediante el algoritmo KMeans, las viviendas se agruparon en función de los atributos de superficie y precio.
# - Los resultados de la agrupación se visualizaron en un gráfico.
```
