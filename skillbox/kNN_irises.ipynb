Vamos a dividir *los datos en conjuntos de datos de entrenamiento y de prueba y*, para simplificar la implementación del algoritmo, combinaremos las matrices de atributos de los objetos y sus etiquetas de clase para dejar claro a qué clase pertenece cada objeto.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(iris_dataset.data[:, 2:4], 
                                                    iris_dataset['target'],
                                                    random_state=0) # random_state - para la reproducibilidad

print(f'X_train shape: {x_train.shape}, y_train shape: {y_train.shape},\n'
      f'X_test shape: {x_test.shape}, y_test shape: {y_test.shape}')
```

```python
import numpy as np

x_train_concat = np.concatenate((x_train, y_train.reshape(112, 1)), axis=1) # Esta llamada a la función concatenar combina las matrices x_train e y_train.reshape(112, 1) a lo largo del segundo eje (horizontalmente).
x_test_concat = np.concatenate((x_test, y_test.reshape(38, 1)), axis=1)
print(f'X_train shape: {x_train_concat.shape},\n'
      f'X_test shape: {x_test_concat.shape}')
```

```python
pd.DataFrame(x_train_concat).head(5)
```

en la 3ª columna tenemos las etiquetas de clase.

Pasemos a la aplicación del algoritmo.

Por defecto, el algoritmo utiliza la métrica de Minkowski.

```python
import math # para la raíz cuadrada

def euclidean_distance(data1, data2):
    distance = 0
    for i in range (len(data1) - 1): # utilizamos todas las columnas excepto la última, por lo que -1
        distance += (data1[i] - data2[i]) ** 2
    return math.sqrt(distance)
```

Calculemos las distancias a todos los puntos de la muestra de entrenamiento y seleccionemos k vecinos (es decir, aquellos cuyas distancias son mínimas).

```python
def get_neighbors(train, test, k=1):

    # Calcular la distancia euclidiana entre el ejemplo de prueba y todos los ejemplos de entrenamiento
    distances = [(train[i][-1], euclidean_distance(train[i], test))
                  for i in range (len(train))]
    distances.sort(key=lambda elem: elem[1]) # Ordenar la lista de distancias en orden ascendente
    #Seleccionamos los k primeros elementos (vecinos) de la lista ordenada
    neighbors = [distances[i][0] for i in range (k)]
    return neighbors
```

Obtengamos ahora una predicción basada en las clases vecinas. Contemos cuántos objetos de cada clase están presentes entre los k más cercanos al objetivo y, a continuación, asignémoslo a la clase cuyas instancias sean más numerosas.

```python
def prediction(neighbors):
    count = {}
    for instance in neighbors:
        if instance in count:
            count[instance] +=1
        else :
            count[instance] = 1

    # Encontramos la etiqueta de clase que aparece con más frecuencia entre los vecinos        
    target = max(count.items(), key=lambda x: x[1])[0]
    return target
```

Escribamos la última función para evaluar la precisión de las previsiones. 

```python
def accuracy(test, test_prediction):
    correct = 0
    for i in range (len(test)):
        if test[i][-1] == test_prediction[i]:
            correct += 1
    return (correct / len(test))
```

```python
predictions = []
for x in range (len(x_test_concat)):
    neighbors = get_neighbors(x_train_concat, x_test_concat[x], k=5)
    result = prediction(neighbors)
    predictions.append(result)
#     print(f'predicted = {result}, actual = {x_test_concat[x][-1]}') # si está interesado en ver qué predicciones concretas son incorrectas.
accuracy = accuracy(x_test_concat, predictions)
print(f'Accuracy: {accuracy}')
```
